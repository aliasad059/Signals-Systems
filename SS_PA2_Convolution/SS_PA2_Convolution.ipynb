## 1. Signal plotting

```python
def plot_dt_signal(n, x_n):
    plt.grid(ls='--')
    plt.axhline(y=0, color='k', linewidth=1)
    plt.axvline(x=0, color='k', linewidth=1)
    plt.rcParams["figure.figsize"] = (15, 10)
    
    input_size = len(n)
    for i in range(input_size):
        plt.plot([n[i], n[i]], [0, x_n[i]], c='tab:blue')
    plt.scatter(n, x_n, c='tab:blue')
    pass
```

```python
def plot_ct_signal(t, x_t):
    plt.grid(ls='--')
    plt.axhline(y=0, color='k', linewidth=1)
    plt.axvline(x=0, color='k', linewidth=1)
    plt.rcParams["figure.figsize"] = (15, 10)
    
    plt.plot(t, x_t, c='tab:blue')
    pass
```

## 2. Compute Convolution

```python
def convolve(x, h):
    n = len(x)
    y = np.zeros(2*n)
    h = h[::-1]
    for i in range(n): 
        y[i]   = np.sum(h[n-i-1:] * x[:i+1])
        y[n+i] = np.sum(h[:n-i-1] * x[i+1:])
    return y[:2*n-1]
```

```python
n = np.arange(-20, 81)
x = np.heaviside(n+2, 1) - 2*np.heaviside(n-12, 1) + np.heaviside(n-20, 1)
h = 0.9**n*(np.heaviside(n-2, 1) - 2*np.heaviside(n-4, 1))
y = convolve(x, h)
n = np.arange(-40, 161)
plot_dt_signal(n,y)
```
